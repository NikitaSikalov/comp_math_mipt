## Построение сплайна

**Задача 1:** выведите систему уравнений для значений 1-х производных свободного кубического сплайна в узлах.

**Задача 2:** Возьмите произвольный отрезок $[a,b]$ и произвольный кубический многочлен, вычислите значения его 1-й и 2-й производных на концах отрезка. С помощью готовой функции (пример ниже) создайте по 2-м точкам кубический сплайн, задав такие же граничные условия. Убедитесь, что сплайн совпадает с исходным многочленом.

```python
%matplotlib inline
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
f = lambda x: 1 / (25*x**2 + 1)
n = 10
x = np.linspace(-1,1,10)
y = f(x)
cs = CubicSpline(x, y, bc_type = 'natural') # Нужно поменять г.у. здесь
xs = np.linspace(-1.1,1.1,100)
plt.figure(figsize=(6.5, 4))
plt.plot(x, y, 'o', label='data')
plt.plot(xs, f(xs), label='true')
plt.plot(xs, cs(xs), label="S")
plt.xlim(-1, 1)
plt.legend()
```
