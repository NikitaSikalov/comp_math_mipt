## Векторные нормированные пространства

Для работы с векторами из $\mathbb{R}^n$ необходимо ввести аналог модуля $|\cdot|$, чтобы перенести весь аппарат связанный с пределами из $\mathbb{R}^1$

* Пусть $V$ - вещественное или комплексное векторное пространство, $f(x)$ - неотрицательная функция, определенная на элементах $x \in V$. Функция называется нормой, если она обладает следующими свойствами:

    1. $f(x) \ge 0 \; \forall x \in V$; $f(x) = 0 \Leftrightarrow x = 0$;
    2. $f(\alpha x) = |\alpha|f(x)$, $\forall \alpha \in \mathbb{R}$ ( или $\mathbb{C})$
    3. $f(x+y) \le f(x) + f(y)$ (неравенство треугольника)

* В любом нормированном пространстве можно ввести метрику $\rho(x,y) = ||x-y||$.

* Пространство называется полным, если любая фундаментальная последовательность сходится. 
* Полное нормированное пространство называется *банаховым*.

## Популярные векторные нормы 

**Пространство векторов из $\mathbb{C}^n$ (или $\mathbb{R}^n$)**

$\forall p \ge 1$ функция

$\displaystyle ||x||_p = \left(\sum_{i=1}^{n} |x_i|^p \right)^{1/p}$

является нормой и называется "$p$-норма вектора". К $p$-нормам относят также 

$\displaystyle ||x||_\infty = \lim_{p \to \infty} ||x||_p = \max_i|x_i| $

Доказательство неравенства треугольника для $p$-нормы сводится к доказательству неравенства Минковского, которое доказывается с помощью неравенства Гёльдера.

## Эквивалентность норм

Несмотря на разное численное значение норм, все они эквивалентны с точки зрения сходимости.

**Определение.** Нормы $||\cdot||_{(1)} и ||\cdot||_{(2)}$ заданные на векторном пространстве $V$ называются эквивалентными, если $\exists c_1, c_2$:

$c_1 ||x||_{(1)} \le ||x||_{(2)} \le c_2 ||x||_{(1)} \; \forall x \in V$

**Теорема.** Любые две нормы в конечномерном векторном пространстве эквивалентны.

## Матричные нормы

* Матрицы одинакового размера составляют конечномерное векторное пространство. Поэтому любая норма конечномерного векторного пространства может использоваться как норма матрицы. Но *матричная норма* удовлетворяет ещё одному условию:

**Определение** $||A||$ называется матричной нормой, если:

1. на векторном пространстве матриц одинакового размера $||A||$ является векторной нормой
2. для любых матриц $A$ и $B$ допускающих умножение выполняется: $\displaystyle ||AB|| \le ||A|| \;||B||$ (мультипликативность)

* Пример матричной нормы - *норма Фробениуса*: 

$\displaystyle ||A||_F = \left(\sum_{i=1}^m \sum_{j=1}^n |a_{ij}|^2\right)^{1/2}$


## Матричные нормы

Доказательство выполнения свойства 2 для нормы Фробениуса:

$A = [a_1, \ldots, a_n] \in \mathbb{C}^{m \times n}$ и 
$$
B = \left[
\begin{array}{c} 
b_1^T & \ldots b_n^T 
\end{array}
\right]
$$
Тогда 
$$AB = \sum_{i=1}^n a_i b_i^T$$

$$
\begin{align}
\Vert AB \Vert_F & \le  \sum_{i} \Vert a_i b_i^T \Vert_F \mbox{ (нер-во треугольника) }\\
& = \sum_{i}  \Vert a_i \Vert_2  \Vert b_i \Vert_2 \mbox{ (следует из формулы для нормы Фробениуса)}\\
& \le \left( \sum_{i} \Vert a_i \Vert_2^2 \right)^{1/2} \left( \sum_{i} \Vert b_i \Vert_2^2 \right)^{1/2} = \Vert A \Vert_F \Vert B \Vert_F \mbox{ (неравенство Коши-Буняковского)}
\end{align}
$$

## Операторы, операторные нормы

* Пусть $X, Y$ - метрические пространства, $D \subset X$. 

* Если каждому элементу $x \in D$ поставлен в соответствие единственный элемент $y \in Y$, то говорят, что на $D$ задан оператор $P$: $y = Px$ (или $y = P(x)$). 

* Оператор является обобщением понятия функции. Можно напрямую перенести с функций на операторы различные определения, например: непрерывность, Липшиц-непрерывность (сделайте это). 

* В вычислительной математике чаще встречаются *линейные операторы в нормированных пространствах*.

** Определение (норма оператора)**

Пусть $P$ действует из $X$ в $Y$ (нормированные пространства). Тогда 

$\displaystyle||P|| = \sup_{x \ne 0} \frac{||Px||_Y}{||x||_X}$

**Докажите, что $\Vert \cdot \Vert_F$ не является операторной нормой.**

## Матричные нормы, порожденные векторными

* Матрица размера $m \times n$ соответствует линейному оператору, действующему из пространства $\mathbb{C}^n$ в пространство $\mathbb{C}^m$, норма этого оператора определяется нормами в этих пространствах:

    * $\displaystyle ||A|| = \max_{x \ne 0}\frac{||Ax||_{(2)}}{||x||_{(1)}}$

* Любая операторная норма матрицы будет матричной нормой в смысле предыдущего определения. 

* В частности, для $p-норм$ при $p=1, 2, \infty$ получаются такие матричные нормы:

    * $\displaystyle ||A||_1 = \max_{1 \le j \le n} \sum_{i = 1}^m |a_{ij}|$ (максимальная сумма модулей по столбцу)

    * $\displaystyle ||A||_\infty = \max_{1 \le i \le m} \sum_{j=1}^n |a_{ij}|$ (максимальная сумма модулей по строке)

    * $||A||_2 = \max_k \sigma_k$ (равна максимальному сингулярному числу матрицы $A$).

## Скалярное произведение

Пусть $V$ - вещественное или комплексное векторное пространство, на котором задана билинейная форма, т.е. для каждой пары векторов задано число $(x, y)$ такое, что:

1. $(x,x) \ge 0$, $\forall x$, $(x,x) = 0 \Leftrightarrow x = 0$
2. $(x,y) = \overline{(y,x)}$
3. $(\alpha x, y) = \alpha (x,y)$, $\alpha$ - число
4. $(x + y, z) = (x,z) + (y,z)$

Тогда $(x,y)$ называется *скалярным произведением* векторов $x$ и $y$. 

* Вещественное пространство со скалярным произведением называется *евклидовым*. 

* Комплексное пространство со скалярным произведеним называется *унитарным*. 

* Стандартное с.п. в $\mathbb{C}^n: \; (x,y) = \sum x_k \bar{y}_k$ порождает 2-норму: $\Vert x \Vert_2 = (x,x)^{1/2}$

## Свойства скалярного произведения
* Скалярное произведение позволяет определить *длину* вектора $x$ как $\sqrt{(x,x)}$. Длина вектора является нормой. Это следует из неравенства *Коши-Буняковского-Шварца*:  
    $\displaystyle |(x,y)| \le (x,x)^{1/2} (y,y)^{1/2}$

* Равенство достигается тогда и только тогда, когда $x$ и $y$ линейно зависимы (докажите).

* Для нормы, порождённой скалярным произведением, выполняется *тождество параллелограмма*:

    * $||x+y||^2 + ||x-y||^2 = 2||x||^2 + 2||y||^2$ (сумма квадратов диагоналей равна сумме квадратов сторон)

* **Теорема.** Для того, чтобы норма на векторном пространстве порождалась скалярным произведением, необходимо и достаточно, чтобы она удовлетворяла неравенству параллелограмма. 

* Поэтому, если для какой-то нормы это неравенство не выполняется, нельзя подобрать скалярное произведение, которое порождало бы эту норму. Например, норма $||\cdot||_1$ не порождается никаким скалярным произведением.

## Изометричные матрицы 

* **Определение:** Матрица $Q \in \mathbb{C}^{n \times n}$  называется *сохраняющей норму* $\Vert \cdot \Vert$, или *изометричной* относительно нормы $\Vert \cdot \Vert$ на $\mathbb{C}^n$, если 
$$ \Vert Q x \Vert = \Vert x \Vert \; \forall x \in \mathbb{C}^n$$
* При $p \ne 2$ матрицы, сохраняющие $p$-норму имеют вид:
$$ Q = P \times diag(d_1, \ldots, d_n)$$
$P$ - матрица перестановки, образована из единичной перестановкой строк.

* При $p = 2$ множество изометричных матриц шире. Сохранение $2$-нормы влечет сохранение скалярного произведения:  
    $(Qx, Qy) = (x,y)$
* Подставляя вместо $x,y$ столбцы единичной матрицы находим:
    $q_i^* q_j = \delta_{ij} \Leftrightarrow Q^* Q = I$, т.е. $Q^* = Q^{-1}$
* **Определение:** матрица $Q \in \mathbb{C}^{n \times n}$ : $Q^* = Q^{-1}$ называется *унитарной*.
* Только унитарные матрицы сохраняют $2$-норму
* Обратная к унитарной - унитарная; произведение унитарных - унитарная матрица.

## Теорема Шура (разложение Шура)

* **Теорема:** $\forall$ $A \in \mathbb{C}^{n \times n}$ с собственными значениями $\lambda_1, \ldots, \lambda_n$ $\exists$ унитарная $Q$ такая, что:
    * Матрица $T = Q^* A Q$ - верхняя треугольная
    * На диагонали - собственные числа: $diag(T) = [\lambda_1, \lambda_2, \ldots, \lambda_n]$
* Доказательство:
    * Пусть $A v_1 = \lambda_1 v_1$, $\Vert v_1\Vert_2 = 1$
    * Дополним $v_1$ до ортонормированного базиса, получим унитарную матрицу: $V_1 = [v_1|v_2|\ldots|v_n]$. Тогда:
    $$
    \begin{align}
    V_1^* A V_1 =
    \left[
    \begin{array}{cccc}
        \lambda_1 & \times & \ldots & \times \\
        0 & & & \\
        \ldots & & A_1 & \\
        0 & & & 
    \end{array}\right]
    \end{align}
    $$
    * Далее: тоже самое для матрицы $A_1$ и т.д. (по индукции)

## Нормальные матрицы

* **Определение:** Матрица называется *нормальной*, если $A^* A = A A^*$
* Классы нормальных матриц:
    * Эрмитовы матрицы: $H^* = H$
    * Унитарные матрицы: $U^* = U^{-1}$  
* В вещественном случае эрмитова матрица называется *симметричной*, унитарная - *ортогональной*.

## Диагонализация нормальной матрицы 

**Теорема** Матрица $A \in \mathbb{C}^{n \times n}$ является нормальной тогда и только тогда, когда в $\mathbb{C}^n$ $\exists$ ортонормированный базис из ее собственных векторов.  
**Доказательство:**  
* ($\Leftarrow$): если существует ортонормированный базис из правых собственных векторов $U = [u_1|\ldots|u_n]$, то $A U = U \Lambda$, где $U$ - унитарная: $U^* = U^{-1}$ $\Rightarrow A = U \Lambda U^*$:
$$
\begin{align}
& A^* A = (U \Lambda U^*)^* \; (U \Lambda U^*) = U \Lambda^* U^* U \Lambda U^* = U \Lambda \Lambda^* U^* = U |\Lambda|^2 U^* \\
& A A^* = (U \Lambda U^*) \; (U \Lambda U^*)^* = U \Lambda U^* U \Lambda^* U^* = U \Lambda^* \Lambda U^* = U |\Lambda|^2 U^* 
\end{align}
$$
* ($\Rightarrow$): 
    * по теореме Шура $A = U T U^*$, где $T$ - верхнетреугольная. 
    * из $A^* A = A A^*$ следует, что $T^* T = T T^*$. Это верно, только если $T$ - диагональная.
    * Следовательно, $T = \Lambda$ и из разложения Шура получаем, что $A U = U \Lambda$, т.е. столбцы $U$ образуют базис из собственных векторов $A$



## Знакоопределенные матрицы
* Следствия из теоремы $A = U \Lambda U^*$
    * Нормальная матрица является эрмитовой $\Leftrightarrow$ все собственные числа $\in \mathbb{R}$.
    * Нормальная матрица является унитарной $\Leftrightarrow$ все собственные числа по модулю равны $1$.

* Любая матрица $A \in \mathbb{C}^{n \times n}$ однозначно представима в виде: $A = H + i K$.   
$H = (A + A^*)/2$, $K = i(A^* - A)/2$ - эрмитовы. 

* На множестве эрмитовых матриц можно ввести понятие знакоопределенности:  
**Определение:** Если $\forall x \in \mathbb{C}^n \ne 0: \; (Ax, x) > 0$, то матрица $A$ называется *положительно определенной*.
* Простой факт: $A^*A \ge 0$: $(A^*Ax,x) = x^* (A^*A x) = (x^*A^*) (Ax) = (Ax)^* (Ax) = \Vert Ax \Vert_2^2 \ge 0$


## Сингулярное разложение

**Теорема** Для любой $A \in \mathbb{C}^{m \times n}$ ранга $r$ существует представление:
$$
A = V\Sigma U^*
$$
где $U \in \mathbb{C}^{n \times n}$, $V \in \mathbb{C}^{m \times m}$ - унитарные, $\Sigma \in \mathbb{C}^{m \times n}$ - диагональная матрица вида:
$$
\Sigma = 
\left[
\begin{array}{cccc}
\sigma_1 & & & \\
 &  \ddots & & \\
 & & \sigma_r & \\
 & & & 
\end{array}
\right]
$$

## Сингулярное разложение, доказательство
* Матрица $A^*A \ge 0$ размера $n \times n$ - эрмитова $\Rightarrow$ $\exists$ унитарная $U = [u_1|\ldots|u_n] \in \mathbb{C}^{n \times n}$:
$$
U^* (A^* A) U  = diag(\sigma_1^2, \ldots,\sigma_n^2), \; \sigma_1 \ge \sigma_2 \ge \ldots \ge \sigma_n
$$
* Обозначим число ненулевых $\sigma_i$ через $r$, т.е.  $\sigma_i = 0$ при $i > r$. Возьмем "урезанные" матрицы $U_r = [u_1, \ldots, u_r]$ и $\Sigma_r = diag(\sigma_1, \ldots, \sigma_r)$, тогда:
$$
U^*_r A^* A U_r = \Sigma_r^2 \Rightarrow (\Sigma_r^{-1} U_r^* A^*) (A U_r \Sigma_r^-1) = I_{r\times r}
$$
* Мы получили матрицу $V_r = A U_r \Sigma_r^{-1}$ размера $m \times r$, для которой $V^*_r V_r = I_{r \times r}$ $\Rightarrow$ столбцы $V_r$ - ортонормированные.
* При $1 \le i \le r$: $A u_i = \sigma_i v_i$. При $i >r$ $A^* A u_i = 0 \Rightarrow (u_i^* A^*) (A u_i) = \Vert A u_i \Vert_2 =0 \Rightarrow Au_i = 0$.
* Достроим $V_r$ до унитарной матрицы $V \in \mathbb{C}^{m \times m}$, получим
$$
A U  = V 
\left[
\begin{array}{cc}
\Sigma_r & 0\\
0 & 0
\end{array}
\right] 
\Rightarrow
V^* A U = 
\left[
\begin{array}{cc}
\Sigma_r & 0\\
0 & 0
\end{array}
\right] 
\Rightarrow
A = V \Sigma_{m\times n} U^*
$$
* Т.к. умножение на невырожденные матрицы не меняет ранг, то $rank(A) = rank(\Sigma) = r$

## Сингулярное разложение, следствия

* Сингулярное разложение удобно записывать в таком виде:

$A = \displaystyle{\sum_{\alpha=1}^r} \sigma_\alpha u_\alpha v_\alpha^*$
или поэлементно:

$a_{ij} = \displaystyle{\sum_{\alpha=1}^r} \sigma_\alpha u_{i\alpha} \overline{v}_{j\alpha}$

Это похоже на разделение переменных в непрерывном случае: мы представляем функцию двух индексов в виде суммы произведение функций одного индекса.

* Сингулярное разложение позволяет найти 4 основных подпространства, связанных с матрицей $A$:

$$\text{ker}(A) = \mathcal{L}\{v_{r+1},\dots,v_n\}$$
$$\text{im}(A) = \mathcal{L}\{u_{1},\dots,u_r\}$$
$$\text{ker}(A^*) = \mathcal{L}\{u_{r+1},\dots,u_n\}$$
$$\text{im}(A^*) = \mathcal{L}\{v_{1},\dots,v_r\}$$

## SVD и приближение малого ранга

Наилучшее приближение матрицы матрицей малого ранга вычисляется через SVD: 

**Теорема(Эккарта-Янга)** Пусть $r < \text{rank}(A)$, $A_r = U_r \Sigma_r V_r^*$. Тогда

$$
    \min_{\text{rank}(B)=r} \|A - B\|_2 = \|A - A_r\|_2 = \sigma_{r+1}.
$$

Для фробениусовой нормы:$\|A - A_r\|_F = \sqrt{\sigma_{r+1}^2 + \dots + \sigma_{\min (n,m)}^2}$.

**Доказательство**

Т.к. $\text{rank} (B) = r$, $\text{dim}~\text{ker}~B = n-r$.<br>
Следовательно $\exists$ $z \ne 0$ такой, что $z\in \text{ker}(B) \cap \mathcal{L}(v_1,\dots,v_{r+1})$.<br>
Возьмем $\|z\| = 1$. Тогда:
$$
    \|A-B\|_2^2 \geq \|(A-B)z\|_2^2 = \|Az\|_2^2 = \| U\Sigma V^* z\|_2= \|\Sigma V^* z\|_2 = \sum_{i=1}^{n} \sigma_i^2 (v_i^*z)^2 =\sum_{i=1}^{r+1} \sigma_i^2 (v_i^*z)^2 \geq \sigma_{r+1}^2\sum_{i=1}^{r+1} (v_i^*z)^2 = \sigma_{r+1}^2
$$
так как $\sigma_1\geq \dots \geq \sigma_{r+1}$ и 
$$\sum_{i=1}^{r+1} (v_i^*z)^2 = \|Vz\|_2^2 = \|z\|_2^2 = 1$$

## Пример. Значения функции на 2-мерной сетке

```python
%matplotlib inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
n = 128
t = np.linspace(0, 1, n)
x, y = np.meshgrid(t, t)
f = 1/(x**2 + y**2 + 1)#(np.sin(np.sin(x/(y+1)))) * np.exp(x*y**2)/(1+x**y)
u, s, v = np.linalg.svd(f, full_matrices=False)
S = s
r = 15
u = u[:, :r]
s = s[:r]
v = v[:r, :] # Транспонированная матрица
fappr = u.dot(np.diag(s).dot(v))
er = np.linalg.norm(fappr - f, 'fro') / np.linalg.norm(f, 'fro')
print('error = ', er)
plt.semilogy(S, 'ko')
```

## Пример 2. Сжатие изображений

Цифровое изображение задаётся 3-мя матрицами (Read Green Blue), сжимая каждую из них с помощью SVD можно сжать исходное изображение.

```python
from scipy import ndimage
import imageio

face = imageio.imread('figures/fig1.jpg') # Читаем файл в многомерный массив 
print(face.shape, face.dtype) # размеры и тип
face_appr = np.zeros(shape = face.shape, dtype = float) # создаём массив для приближенного изображения
r = 6 #max(face.shape[0:1]) # сколько сингулярных чисел мы оставим 
for i in range(0,3):
    u, s, v = np.linalg.svd(face[:,:,i])
    u_r = u[:, :r]
    s_r = s[:r]
    v_r = v[:r, :] # Mind the transpose here!
    face_appr[:,:,i] = u_r.dot(np.diag(s_r).dot(v_r))
face_appr = np.rint(face_appr)
face_appr = face_appr.astype('uint8')
error = face-face_appr
print('average error = ', (np.sum(error**2)/(face.shape[0] * face.shape[1] * 3))**(0.5), '\n max error = ', np.max(error))
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,20))
ax1.imshow(face_appr) 
ax2.imshow(face)
# plt.savefig('svd_of_image.png', dpi = 300, format = 'png')
```

## Теория возмущений

* Пусть $x$ вычисляется по $f(x)$. Как сильно меняется $f(x)$ при малых изменениях $x$?
* Если $f$ - функция, $f(x + \delta) \approx f(x) + f' \delta$. Если $f(x)$ и $x$ $\ne 0$
$$
\Vert \frac{f(x+\delta) - f(x) \Vert}{\Vert f(x) \Vert} \approx \left(\frac{ \Vert f'(x) \Vert}{\Vert f(x)\Vert} \Vert x \Vert \right) \frac{\Vert \delta \Vert}{\Vert x \Vert}
$$
* Мерой чувствительности задачи (как относительная ошибка зависит от относительного возмущения) является *число обусловленности* (*condition number*)
$$
cond(f(x)) = \frac{ \Vert f'(x) \Vert}{\Vert f(x)\Vert} \Vert x \Vert
$$
* Какое число обусловленности для вычисления обратной матрицы?

## Число обусловленности матрицы

* Пусть $A$ - невырожденная матрица, тогда 
$$
\begin{align}
(A + \Delta)^{-1} - A^{-1} = -A^{-1} \Delta (A + \Delta)^{-1} \approx - A^{-1} \Delta A^{-1} \Rightarrow \\
\frac{\Vert (A + \Delta)^{-1} - A^{-1} \Vert}{\Vert A^{-1} \Vert} \le \approx \left( \Vert A^{-1} \Vert \Vert A \Vert \right) \frac{\Vert \Delta \Vert}{\Vert A \Vert}
\end{align}
$$
* Величину
$$
cond(A) = \Vert A^{-1} \Vert \Vert A \Vert
$$
называют *числом обусловленности матрицы* $A$. 

* В выкладках мы предполагали, что $(A + \Delta)^{-1} \approx A$ при малых $\Delta$. Нужно выяснить, при каких $\Delta$ матрица $(A + \Delta)$ будет невырожденной.

### Операторные ряды
* Ряд $\displaystyle \sum_{k=0}^\infty A_k$, где $A_k \in \mathbb{C^{n \times n}}$ называется сходящимся, если сходится последовательность его частичных сумм $\displaystyle S_N = \sum_{k=0}^N A_k$.

* Для сходимости достаточно, чтобы сходился числовой ряд $\displaystyle \sum_{k=0}^\infty \Vert A_k \Vert$ (докажите).

* Ряд $\displaystyle \sum_{k=0}^\infty F^k$ называется *рядом Неймана*. Достаточное условие сходимости: $\Vert F \Vert < 1$ (для любой нормы!).

* Максимальное по модулю собственное число матрицы  называется её *спектральным радиусом* $\rho(F)$. Если $\rho(F) < 1$ матрица называется *сходящейся*.   
**Теорема:** Ряд Неймана с матрицей $F$ сходится $\iff$ $\rho(F) < 1$, т.е. матрица $F$ - сходящаяся.

### Операторные ряды
* **Лемма** Если $\Vert F \Vert < 1$, то $A = I - F$ обратима и 

    * $\displaystyle (I-F)^{-1} = \sum_{k = 0}^\infty F^k$

    * $\Vert (I - F)^{-1} \Vert \le \frac{\Vert I \Vert}{1 - \Vert F \Vert}$

(доказывается умножением частичных сумм на $(I - F)$)

* **Следствие:** Если $A$ - невырожденная и $\Vert A^{-1} E \Vert < 1$, то
    * $A + E$ - невырожденная и 
$$
\begin{align}
(A + E)^{-1} = \sum_{k=0}^\infty (-A^{-1}E)^k A^{-1} = A^{-1} \sum_{k=0}^\infty (-E A^{-1})^k
\end{align}
$$
    * 
$$
\frac{\Vert (A + E)^{-1} - A^{-1} \Vert}{\Vert A^{-1} \Vert} \le \frac{\Vert A^{-1}\Vert \Vert E \Vert}{1 - \Vert A^{-1} E \Vert}
$$


## Обусловленность линейной системы

* Рассмотрим линейную систему: $Ax = f$, $f \ne 0$
* Возмущенная система: $(A + \Delta A) \tilde{x} = f + \Delta f$
* Как возмущение решения зависит от возмущения матрицы и правой части?

Пусть $\Vert A^{-1} \Delta A \Vert < 1$
Тогда 
$$
\begin{align}
\Delta x = \tilde{x} - x & = (A + \Delta A)^{-1} (f + \Delta f) - A^{-1} f \\
& = \left[ (A + \Delta A)^{-1} - A^{-1}\right] f + (A + \Delta A)^{-1}\Delta f \\
& = \left[ \sum_{k = 1}^\infty (- A^{-1} \Delta A)^{k} \right] (A^{-1}f) + \left[ \sum_{k=0}^\infty (-A^{-1} \Delta A)^k\right] (A^{-1} \Delta f)
\end{align}
$$
Возьмем норму обеих частей, воспользуемся нер-вом тр-ка и тем, что: $\Vert \sum_{k = 1}^\infty (- A^{-1} \Delta A)^{k} \Vert \le 1/(1-\Vert A^{-1} \Delta A\Vert)$:
$$ \boxed{
\frac{\Vert \Delta x \Vert}{\Vert x \Vert} \le \frac{\Vert A^{-1} \Vert \Vert A \Vert}{1 - \Vert A^{-1} \Delta A \Vert} \left( \frac{\Vert \Delta A \Vert}{\Vert A \Vert } + \frac{\Vert \Delta f \Vert}{\Vert f \Vert} \right)}
$$


```python
# Матрица Вандермонда для равномерной сетки
import numpy as np

N = 30
x = np.linspace(0, 1, N)
V = np.vander(x)
print(np.linalg.cond(V))

c = np.ones(N)
b = np.dot(V, c)

c_ = np.linalg.solve(V, b)

print(np.linalg.norm(c - c_, ord = np.Inf))
```

# Согласованность матрицы и правой части

* Предыдущая оценка верна для любой правой части $f$
* При разных $f$ и $\Delta f$ обусловленность может быть разной  
Можно оценить возмущения во 2-й норме с помощью SVD. В общем случае $\Vert A \Vert \Vert A^{-1} \Vert = \sigma_1/\sigma_n$.
* Пусть $A = V \Sigma U^* = \sum_k \sigma_k v_k u_k^*$.
* Пусть $f = \sum_k a_k v_k$, $\Delta f = \sum_k b_k v_k$
* Тогда $x = \sum_k (a_k/\sigma_k)u_k$, $\Delta x = \sum_k (b_k/\sigma_k) u_k$
* Пусть $f$ и $\Delta f$ имеют нулевые составляющие в направление младших сингулярных векторов $v_{r+1},\ldots, v_{n}$ (*говорят, что матрица и правая часть согласованы*)
* Следовательно:
 $$
 \begin{align}
 & \frac{\Vert \Delta x \Vert_2}{\Vert x \Vert_2} = \frac{\left(\sum_{k=1}^{r} (b_k/\sigma_k)^2\right)^{1/2}}{\left(\sum_{k=1}^{r} (a_k/\sigma_k)^2\right)^{1/2}} , \;\frac{\Vert \Delta f \Vert_2}{\Vert f \Vert_2} = \frac{\left(\sum_{k=1}^{r} (b_k)^2\right)^{1/2}}{\left(\sum_{k=1}^{r} (a_k)^2\right)^{1/2}} \Rightarrow\\
 & \boxed{\frac{\Vert \Delta x \Vert_2}{\Vert x \Vert_2} \le \frac{\sigma_1}{\sigma_r} \frac{\Vert \Delta f \Vert_2}{\Vert f \Vert_2}}
 \end{align}
 $$

## Диагональное преобладание

* Матрица $A \in \mathbb{C}^{n \times n}$ имеет *строчное диагональное преобладание*, если 
$$
|a_{ii}| > r_i \equiv \sum_{j=1,\; j \ne i}^{n} |a_{ij}|, \; i = 1, \ldots, n
$$
и *столбцовое диагональное преобладание*, если 
$$
|a_{jj}| > c_j \equiv \sum_{i=1,\; i \ne j}^{n} |a_{ij}|, \; j = 1, \ldots, n
$$

* **Теорема:** Матрица, имеющая диагональное преобладание, является невырожденной.
    * обозначим $\mbox{diag}(A) = \mbox{diag}(a_{11}, \ldots, a_{nn})$ диагональную часть матрицы A и  
    $\mbox{off}(A) = A - \mbox{diag}(A)$ - внедиагональную часть
    * Тогда для строчного д.п. $\Vert [\mbox{diag}(A)]^{-1} \mbox{off}(A)\Vert_\infty < 1$, следовательно  
    матрица $I + [\mbox{diag}(A)]^{-1} \mbox{off}(A)$ - невырожденная (доказали ранее)
    * Домножим эту матрицу на невырожденную $\mbox{diag}(A)$ получим $\mbox{diag}(A) + \mbox{off}(A) = A $



## Круги Гершгорина
* **Теорема:** Для $A \in \mathbb{C}^{n \times n}$ рассмотрим круги
$$
R_i \equiv \left\{ z \in \mathbb{C}: |a_{ii} - z| \le r_i \right\}, \quad C_i \equiv \left\{ z \in \mathbb{C}: |a_{ii} - z| \le C_i \right\}
$$
где 
$$ r_i \equiv \sum_{j=1,\; j \ne i}^{n} |a_{ij}|, \; i = 1, \ldots, n,\quad  c_j \equiv \sum_{i=1,\; i \ne j}^{n} |a_{ij}|, \; j = 1, \ldots, n$$
Тогда для собственного числа $\lambda$ матрицы $A$ верно, что 
$$
\lambda \in \bigcup_{i=1}^n R_i \mbox{ и } \lambda \in \bigcup_{i=1}^n C_i
$$
    * Если $\lambda \notin \bigcup_i R_i$, то $A -\lambda I$ имеет строчное диагональное преобладание, и следовательно невырождена
    * Это противоречит тому, что $\lambda$ - собственное число $A$
* **Теорема**: Если $m$ кругов Гершгорина образуют область $G$, непересекающуюся с остальными кругами, то в $G$ находятся **ровно** $m$ с.ч.

* С помощью кругов Гершгорина можно получить оценку на границы спектра $A$

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
n = 3
A = np.random.rand(n,n)  +  np.diag(np.arange(n)**2+1) # 1j * np.random.rand(n,n)
l = np.linalg.eigvals(A)

circles = []
fig, ax = plt.subplots(figsize = (10,10))
for i in range(n):
    ax.add_patch(matplotlib.patches.Circle((np.real(A[i,i]), np.imag(A[i,i])), np.sum(np.abs(A[i,:])) - A[i,i], fill = False))
ax.plot(np.real(l), np.imag(l),'ro')
ax.set_aspect('equal')
```
